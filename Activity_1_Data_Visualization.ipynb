```python
# ACTIVITY 1 - DATA VISUALIZATION AND DASHBOARD DEPLOYMENT
# Students: Elis García and Neris Pacheco

# POINT 1

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8,5)

df = pd.read_csv("university_student_data.csv")

# Preview
print("Dataset preview:")
display(df.head())

# General information
print("\nDataset information:")
print(df.info())

# Missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# Descriptive statistics
print("\nDescriptive statistics:")
display(df.describe())

# Column descriptions
columns = {
    'Year': 'Academic year of the record',
    'Department': 'Academic department',
    'Term': 'Academic term (Spring or Fall)',
    'Applications': 'Total number of applications received',
    'Enrollments': 'Number of students enrolled',
    'Retention_Rate': 'Percentage of students who remain in the program',
    'Satisfaction_Score': 'Average student satisfaction score (0–100)'
}

print("\nColumn descriptions:")
for col, desc in columns.items():
    print(f"- {col}: {desc}")

```

```python
# POINT 2

import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8,5)

# a) Trends in retention rate over time
# Group data by year and calculate the average "Retention Rate (%)"
retention_by_year = df.groupby('Year')['Retention Rate (%)'].mean().reset_index()

plt.figure()
sns.lineplot(
    data=retention_by_year,
    x='Year',
    y='Retention Rate (%)',
    marker='o',
    color='steelblue'
)
plt.title("Trend of Retention Rate Over Time")
plt.xlabel("Year")
plt.ylabel("Average Retention Rate (%)")
plt.show()

# b) Student satisfaction scores by year
# Group by year and calculate the average "Student Satisfaction (%)"
satisfaction_by_year = df.groupby('Year')['Student Satisfaction (%)'].mean().reset_index()

plt.figure()
sns.barplot(
    data=satisfaction_by_year,
    x='Year',
    y='Student Satisfaction (%)',
    hue='Year',
    palette='viridis',
    legend=False
)
plt.title("Student Satisfaction Scores by Year")
plt.xlabel("Year")
plt.ylabel("Average Satisfaction Score (%)")
plt.show()

# c) Comparison between Spring and Fall terms
# Group by term (Spring/Fall) and calculate the average "Retention Rate (%)"
retention_by_term = df.groupby('Term')['Retention Rate (%)'].mean().reset_index()

plt.figure()
sns.barplot(
    data=retention_by_term,
    x='Term',
    y='Retention Rate (%)',
    hue='Term',
    palette='pastel',
    legend=False
)
plt.title("Comparison of Retention Rate Between Spring and Fall")
plt.xlabel("Term")
plt.ylabel("Average Retention Rate (%)")
plt.show()
```
